```python

import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
df = pd.read_csv("Subsidy.csv")
df
```

```python
x_train = df[['CostCultivation','CostCultivation2','Production','Yield','RainFall Annual']][10:]
x_train.shape
```

```python
x_test = df[['CostCultivation','CostCultivation2','Production','Yield','RainFall Annual']][:10]
x_test.shape
```

```python
y_train = df.Price[10:]
y_train.shape
```

```python
y_test = df.Price[:10]
y_test.shape
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)

accuracy = reg.score(x_test,y_test)
accuracy
```
